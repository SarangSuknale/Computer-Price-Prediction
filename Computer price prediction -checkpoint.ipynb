Visualize distribution of all the Categorical Predictor variables in the data using bar plots

Categorical Predictors:

'ram', 'screen', 'cd', 'multi'

```python
def PlotForBar (inp, ColToBar):
    %matplotlib inline
    import matplotlib.pyplot as plt
    
    fig, subplot=plt.subplots(nrows=1, ncols=len(ColToBar), figsize=(20,5))
    fig.suptitle('Bar Chart of:'+str(ColToBar))
    
    for ColName, PlotName in zip(ColToBar, range(len(ColToBar))):
        inp.groupby(ColName).size().plot(kind='bar',ax=subplot[PlotName])
    
    
```

```python
PlotForBar(inp=ComputerPricesData, ColToBar=['speed','ram','screen','cd','multi'])
```
